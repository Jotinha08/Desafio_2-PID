```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

type = 'black'
df = pd.read_csv(f'src/data/{type}/vetores_caracteristicas.csv', header=None)

df['objeto'] = df.iloc[:, 1:6].mean(axis=1)
df['fundo'] = df.iloc[:, 6:8].mean(axis=1)

labels = df.iloc[:, 0].unique()
colors = plt.cm.rainbow(np.linspace(0, 1, len(labels)))

for label, color in zip(labels, colors):
    subset = df[df.iloc[:, 0] == label]
    plt.scatter(subset['objeto'], subset['fundo'], s=100, color=color, label=f'{label}h')

plt.xlabel('Objeto')
plt.ylabel('Fundo')
plt.legend()

plt.savefig(f'src/caracteristicas/{type}/espaco_caracteristicas_2d.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

type = 'black'
df = pd.read_csv(f'src/data/{type}/vetores_caracteristicas.csv', header=None)

labels = df.iloc[:, 0]

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

for label, color in zip(labels.unique(), plt.cm.rainbow(np.linspace(0, 1, len(labels.unique())))):
    subset = df[df.iloc[:, 0] == label]
    ax.scatter(subset[2], subset[4], subset[6], s=100, color=color, label=f'{label}h')

ax.set_xlabel('Entropia')
ax.set_ylabel('Contraste')
ax.set_zlabel('Dimensão Fractal')
ax.legend()

plt.savefig(f'src/caracteristicas/{type}/espaco_caracteristicas_3d.png')
plt.show()
```
